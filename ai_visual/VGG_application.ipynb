# VGG

## Load VGG model from keras

```python
from keras.applications.vgg16 import VGG16

model = VGG16()
model.summary()
```

## Load image file

```python
import os
import cv2
from matplotlib import pyplot as plt
import numpy as np

image_dir = "./flick_images"
image_files = os.listdir(image_dir)

images = []
for i in image_files:
    file_name = image_dir + "/" + i;
    images.append(cv2.imread(file_name))
    
plt.imshow(images[-1])
```

## Resize Image

```python
def resize_image(img, x, y):
    resized_img = cv2.resize(img, (x, y))
    return resized_img

resized = resize_image(images[4], 224, 224)
print("resized image shape =", resized.shape)

resized_ex = np.expand_dims(resized, axis=0)
print("resized_ex image shape =", resized_ex.shape)

plt.imshow(np.squeeze(resized_ex))
```

## VGG Prediction

```python
print("resized_ex image shape =", resized_ex.shape)

yhat = model.predict(resized_ex)
```

## Check

```python
from keras.applications.vgg16 import decode_predictions

label = decode_predictions(yhat)
label = label[0][0]
print(f"{label[1]} ({label[2]*100:.2f}%)")
```
