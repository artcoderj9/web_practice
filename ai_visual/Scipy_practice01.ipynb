```python
%matplotlib inline

from scipy import ndimage, misc
from matplotlib import pyplot as plt
import numpy as np

# get face image of panda from misc package
raccoon = misc.face()

# plot
plt.imshow(raccoon)
plt.show()
```

```python
# flip up/down using numpy
flip_down = np.flipud(raccoon)
plt.imshow(flip_down)
plt.show()
```

```python
# rotate using scipy
raccoon_rot = ndimage.rotate(raccoon, 25)
plt.imshow(raccoon_rot)
plt.show()
```

```python
# croping

raccoon_g = misc.face(gray=True)
print(raccoon_g.shape)

lx, ly = raccoon_g.shape
lx_min_idx = lx // 4
lx_max_idx = -lx_min_idx
ly_min_idx = ly // 4
ly_max_idx = -ly_min_idx
print(lx_min_idx, lx_max_idx, ly_min_idx, ly_max_idx)
raccoon_crop = raccoon_g[lx_min_idx: lx_max_idx, ly_min_idx : ly_max_idx]
plt.imshow(raccoon_crop)
plt.show()
```

```python
# blurring
blurred_face = ndimage.gaussian_filter(raccoon, sigma=3)
very_blurred = ndimage.gaussian_filter(raccoon, sigma=5)

plt.imshow(very_blurred)
plt.show()
```

```python
# sharpen
blurred_face_f = ndimage.gaussian_filter(blurred_face, sigma = 1)

alpha = 10
sharpened = blurred_face + alpha * (blurred_face - blurred_face_f)
plt.imshow(sharpened)
plt.show()
```
